```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as cts
import math as m
```

```python
u = np.linspace(-0.5, 0.5, 50)
v = np.linspace(-0.5, 0.5, 30)
U, V = np.meshgrid(u, v)
```

```python
L = 1
t = 1000000000000
```

```python
# u => x ; y => v
def r(u, v):
  n = 1
  x = 0
  y = 0
  con1 = (n*cts.pi/L)
  con2 = (n*cts.pi/L)**2
  con3 = cts.e/cts.m_e
  while n < 10:
    x = x*con3/2 + (-1 * con2)*m.exp(-1*con1*v)*m.cos(con1*u)*con3*t**2/2
    y = y*con3/2 + con2*m.sin(con1*u)*m.exp(-1*con1*v)*con3*t**2/2
    n = n+1
  return x, y
```

```python
r_vec = np.vectorize(r)
X, Y = r_vec(U, V)
```

```python
fig, ax = plt.subplots()
ax.quiver(U, V, X, Y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
