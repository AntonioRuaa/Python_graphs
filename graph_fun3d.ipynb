```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as cts
import math as m
```

```python
L = 1
```

```python

def f(x, y):
  n = 1
  z = 0
  con = n*m.pi/L
  while n < 15:
    z = z + con*m.sin(con*x)*m.exp(-1*con*y)*(1-con)
    n = n+1
  return z
```

```python
x = np.arange(-10, 10, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x,y)
```

```python
f_vec = np.vectorize(f)
Z= f_vec(X, Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
